```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

```python
boston_data = datasets.load_boston()
data_df= pd.DataFrame(data=boston_data.data,columns= boston_data.feature_names)
data_df["TARGET"] = boston_data.target
data_df
```

```python
data_df.describe()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_df.hist(alpha=0.6,figsize=(6,6))
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler() # Z-scaler 物件
X_scaled = scale.fit_transform(data_df)
X_scaled_df = pd.DataFrame(X_scaled,columns=data_df.keys())
X_scaled_df.hist(alpha=0.6,figsize=(6,6))
plt.tight_layout()
```

```python
sns.set_style("darkgrid")
sns.regplot(x=data_df["RM"], y=data_df["TARGET"],)
plt.tight_layout()
```

```python
from sklearn import preprocessing 
sns.set_style("darkgrid")
min_max_scaler = preprocessing.MinMaxScaler()
column_sels = ["CRIM","ZN","INDUS",
 "CHAS","NOX","RM",
 "AGE","DIS","RAD",
 "TAX","PTRATIO","B",
 "LSTAT"]
x = data_df.loc[:, column_sels]
y = data_df["TARGET"]
x = pd.DataFrame(data=min_max_scaler.fit_transform(x),
 columns=column_sels)
fig, axs = plt.subplots(ncols=3, nrows=5, figsize=(12,16))
index = 0
axs = axs.flatten()
for i, k in enumerate(column_sels):
    sns.regplot(y=y, x=x[k], ax=axs[i],x_jitter=True)
#更改 Matplotlib 子圖大小和間距
fig.tight_layout()
```
